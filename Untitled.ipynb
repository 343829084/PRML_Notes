```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.mlab as mlab
from mpl_toolkits.mplot3d import Axes3D
```

```python
def q(x, y):
    g1 = mlab.bivariate_normal(x, y, 1.0, 1.0, -1, -1, -0.8)
    g2 = mlab.bivariate_normal(x, y, 1.5, 0.8, 1, 2, 0.6)
    return 0.6*g1+28.4*g2/(0.6+28.4)

fig = plt.figure(figsize=(10,6))
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = q(X, Y)
#Z = mlab.bivariate_normal(X, Y, 1.0, 1.0, -1, -1, -0.8)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('jet'),linewidth=0, antialiased=True)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
